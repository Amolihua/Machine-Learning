```python
from sklearn.datasets import load_diabetes
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
%matplotlib inline
```

```python
df=load_diabetes()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
dataset['target']=df.target
```

```python
dataset.head()
```

```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg = LinearRegression()
mse = cross_val_score(lin_reg, x, y, scoring='neg_mean_squared_error', cv=5)
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
lin_reg.fit(x, y)
```

```python
lin_reg.predict([[3,2,1,0,0,0,0,0,0,0]])
```

# RIDGE


```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV
ridge = Ridge()


params = {'alpha': [1e-15,1e-10,1e-8,1e-5,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,60,70,80,90,100]}

ridge_regressor = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=5)
ridge_regressor.fit(x, y)

```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

# LASSO

```python
lasso = Lasso(max_iter=100000)
lasso_regressor = GridSearchCV(lasso, params, scoring='neg_mean_squared_error', cv=5)
lasso_regressor.fit(x, y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
mse_train = cross_val_score(lin_reg, x_train, y_train, scoring='neg_mean_squared_error', cv=5)
print(np.mean(mse_train))
```

```python
ridge_regressor2 = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=10)
lasso_regressor2 = GridSearchCV(lasso, params, scoring='neg_mean_squared_error', cv=10)
ridge_regressor2.fit(x_train, y_train)
lasso_regressor2.fit(x_train, y_train)
print("Ridge:")
print(ridge_regressor2.best_params_)
print(ridge_regressor2.best_score_)
print("Lasso:")
print(lasso_regressor2.best_params_)
print(lasso_regressor2.best_score_)
```

```python
from sklearn.metrics import r2_score
y_pred_lasso = lasso_regressor2.predict(x_test)
y_pred_ridge = ridge_regressor2.predict(x_test)
y_pred_lin = lin_reg.predict(x_test)

r2_score1 = r2_score(y_pred_lasso, y_test)
r2_score2 = r2_score(y_pred_ridge, y_test)
r2_score3 = r2_score(y_pred_lin, y_test)

print("R2 Score Lasso:", r2_score1)
print("R2 Score Ridge:", r2_score2)
print("R2 Score Linear Regression:", r2_score3)
```
